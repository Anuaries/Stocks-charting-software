```python
#importing all modules

import pandas as pd
import time
from datetime import datetime
import pandas as pd


#Period for Calculating RSI , start and end date
# start date
dt = datetime(2010, 1, 1)
start_date = int(round(dt.timestamp()))

# end date
dt = datetime(2017, 12, 31)
end_date = int(round(dt.timestamp()))

#method for calculating RSI
def RSI(ohlc, period=14, method="SMA"):
    delta = ohlc["Close"].diff()
    up, down = delta.copy(), delta.copy()
    up[up < 0] = 0
    down[down > 0] = 0
       
    if method == "SMA":
        # using Smooth Moving Average
        gain = up.rolling(period).mean()
        loss = down.abs().rolling(period).mean()
    else:
        # using Exponential Weighted Moving Average
        gain = up.ewm(span=period).mean()
        loss = down.abs().ewm(span=period).mean()
    RS = gain / loss
    ohlc['RSI'] =  pd.Series(100 - (100 / (1 + RS)), name="RSI")
    
    return ohlc

#Stocks in a list
stocks = ['FB','AAPL','AMZN','IBM','GOOGL','MSFT','NAK','O','QCOM','TSLA']

df_RSI =pd.DataFrame(columns=stocks)
 
for s in stocks:
    df = pd.read_csv(f"https://query1.finance.yahoo.com/v7/finance/download/{s}?period1={start_date}&period2={end_date}&interval=1d&events=history&includeAdjustedClose=true")
    df=RSI(df, period=14, method='SMA')
    display(df) 
    df_RSI[s]=df['RSI']

#setting the index to Date   
df_RSI['RSI'] = df['RSI']
df_RSI['Date']=df['Date'] 
df_RSI.set_index("Date",inplace=True)


from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()
    
    



df_RSI.iplot()
    
df_RSI    
   
```
